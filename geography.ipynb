```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset=pd.read_csv('geography.csv')
```

```python
del dataset['Price Change']
dataset
```

```python
dataset=dataset.iloc[:32,:]
dataset
```

Visualizations done in Excel

Segmentation

```python
df=dataset.iloc[:32,1:]
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data_standard=scaler.fit_transform(df)
```

```python
data_standard
```

```python
import sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
clusters = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(data_standard)
    clusters.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,10),clusters, linewidth=2, color="red", marker ="8")
plt.xlabel("Clusters")
plt.xticks(np.arange(1,10,1))
plt.ylabel("Inertia")
plt.title('Elbow Method')
plt.show
```

Initially we are trying to find the optimal K value using the Elbow Method wherein we will be finding the Within Cluster Sum of Square (WCSS) and try to find the point where it rapidly decreases which makes the graph look like an elbow there. The K value corresponding to that point is the optimal K value. We can dlearly see that this elbow is forming at Clusters=3

K=3

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data_standard)

clusters=kmeans.cluster_centers_
print(clusters)
```

```python
segments = kmeans.fit_predict(data_standard)
```

```python
segments
```

```python
dataset= dataset.assign(segment=segments)
dataset
```
