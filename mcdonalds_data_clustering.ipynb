# **Segmentation**


Converting text data to numerical data 

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
text = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']
for elements in text:
    dataset[elements] = le.fit_transform(dataset[elements])
```

```python
dataset
```

Mapping VisitFrequency and Gender to numerical values as required.Since we are only performing clustering, we can map the visitfrequency value as per our convenience as it wont affect the performence of our clustering model. We are just looking for similarity and not concerned with actual values.

```python
scale_mapper_visit = {"Once a month":1, "Every three months":3, "Once a year":3, "Once a week": 4, "Never" : 0, "More than once a week":5}
dataset["VisitFrequency"] = dataset["VisitFrequency"].replace(scale_mapper_visit)
```

```python
scale_mapper_gender = {"Male":1, "Female":0}
dataset["Gender"] = dataset["Gender"].replace(scale_mapper_gender)
```

```python
dataset
```

Checking correlation of the variables to each other and producing a correlation heatmap

Standardizing the numerical values to increase effeciency of model

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data_standard=scaler.fit_transform(dataset)
```

```python
data_standard
```

K-Means clustering 

```python
import sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

Using elbow method to determine the number of clusters
Inertia is the squared distance of each sample in a cluster to its cluster center and sums them up.

```python
clusters = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(data_standard)
    clusters.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,10),clusters, linewidth=2, color="red", marker ="8")
plt.xlabel("Clusters")
plt.xticks(np.arange(1,10,1))
plt.ylabel("Inertia")
plt.title('Elbow Method')
plt.show
```

So, we can say that the ideal number of clusters would be 5

Implementing K-means clustering

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(data_standard)

clusters=kmeans.cluster_centers_
print(clusters)
```

```python
y_prediction = kmeans.fit_predict(data_standard)
print(y_prediction)
```

```python
dataset= dataset.assign(segment=y_prediction)
```
