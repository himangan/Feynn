# Segmentation

Creating a copy of the dataset for pre-processing for segmentation.

```python
df=dataset
```

Converting Categorical Variables to Numeric

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
text = ['Profession','Education','Marrital Status', 'Personal loan','House Loan','Wife Working','Make']
for elements in text:
    df[elements] = le.fit_transform(df[elements])
```

Standardizing the numerical values to increase effeciency of model

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data_standard=scaler.fit_transform(df)
```

```python
data_standard
```

K-Means clustering

K-means clustering is a type of unsupervised learning, which is used when you have unlabeled data (i.e., data without defined categories or groups). The goal of this algorithm is to find groups in the data, with the number of groups represented by the variable K. The algorithm works iteratively to assign each data point to one of K groups based on the features that are provided. Data points are clustered based on feature similarity. The results of the K-means clustering algorithm are:

The centroids of the K clusters, which can be used to label new data. Labels for the training data (each data point is assigned to a single cluster) . The means in the K-means refers to averaging of the data; that is, finding the centroid.

```python
import sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```


As we are not given any information regarding the number of segments to produce we will be using elbow method to determine the number of clusters Inertia is the squared distance of each sample in a cluster to its cluster center and sums them up.
Initially we are trying to find the optimal K value using the Elbow Method wherein we will be finding the Within Cluster Sum of Square (WCSS) and try to find the point where it rapidly decreases which makes the graph look like an elbow there. The K value corresponding to that point is the optimal K value. We can dlearly see that this elbow is forming at Clusters=4

```python
clusters = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(data_standard)
    clusters.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,10),clusters, linewidth=2, color="red", marker ="8")
plt.xlabel("Clusters")
plt.xticks(np.arange(1,10,1))
plt.ylabel("Inertia")
plt.title('Elbow Method')
plt.show
```

So we can say the ideal number of clusters would be 4

Implementing K-Means clustering

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(data_standard)

clusters=kmeans.cluster_centers_
print(clusters)
```

```python
segments = kmeans.fit_predict(data_standard)
```

```python
segments
```

```python
dataset= dataset.assign(segment=segments)
dataset
```

So, here we have the records for each customer and assigned a segment to each of those customers.The customers in a segment have similar features and that information may be leveraged to create targeted promotion campaigns or design new offers schemes for different target segments.
